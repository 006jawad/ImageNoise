```python
import numpy as np 
import pandas as pd 
from PIL import Image
import numpy as np
from skimage.util import random_noise
import os
```

```python
paths = []

for dirname, _, filenames in os.walk('D:\The_Great_Barrier_Reef\\train_images\\video_2\\'):
    for filename in filenames:
        paths.append(os.path.join(dirname, filename))
```

```python
def addNoise(in_path, out_path, type, clip, *args): #gaussian, localvar, poisson, salt, pepper, s&p,
    im = Image.open(in_path)
    # convert PIL Image to ndarray
    im_arr = np.asarray(im)

    # random_noise() method will convert image in [0, 255] to [0, 1.0],
    # inherently it use np.random.normal() to create normal distribution
    # and adds the generated noised back to image
    if type == 'gaussian':
        noise_img = random_noise(im_arr, mode='gaussian', var=args[0], clip=clip) 
        noise_img = (255*noise_img).astype(np.uint8)
        print(f'value of var given {args[0]}')

    if type == 'localvar':
        noise_img = random_noise(im_arr, mode='localvar', local_vars=args[0], clip=clip) 
        noise_img = (255*noise_img).astype(np.uint8)  
        print(f'value of localvar given {args[0]}')

    if type == 'poisson':
        noise_img = random_noise(im_arr, mode='poisson', clip=clip) 
        noise_img = (255*noise_img).astype(np.uint8) 

    if type == 'salt':
        noise_img = random_noise(im_arr, mode='salt', amount=args[0], clip=clip) 
        noise_img = (255*noise_img).astype(np.uint8) 
        print(f'value of amount given {args[0]}') 

    if type == 's&p':
        try:
            noise_img = random_noise(im_arr, mode='s&p', amount=args[0], salt_vs_pepper=args[1], clip=clip) 
            noise_img = (255*noise_img).astype(np.uint8)  
            print(f'value of amount given {args[0]} and value of salt_vs_pepper given {args[1]}')
        except:
            noise_img = random_noise(im_arr, mode='s&p', amount=args[0], clip=clip) 
            noise_img = (255*noise_img).astype(np.uint8)
            print(f'value of amount given {args[0]}')

    if type == 'speckle':
        try:
            noise_img = random_noise(im_arr, mode='speckle', mean=args[0], var=args[1], clip=clip) 
            noise_img = (255*noise_img).astype(np.uint8)  
            print(f'value of mean given {args[0]} and value of var given {args[1]}')
        except:
            noise_img = random_noise(im_arr, mode='speckle', mean=args[0], clip=clip) 
            noise_img = (255*noise_img).astype(np.uint8)  
            print(f'value of mean given {args[0]}')


    img = Image.fromarray(noise_img)
    img.save(out_path)
    return print(f'Done adding {type} noise....')
```
